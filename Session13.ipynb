```python
import pandas as pd
column_names = ['age','workclass','fnlwgt','education','educationnum','maritalstatus','occupation','relationship','race','sex','capitalgain','capitalloss','hoursperweek','nativecountry','fnlwgt1']
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data",header = None,names = column_names)
df.head()
```

```python
df.drop(df.index[20:],inplace=True)
df
```

```python
import sqlalchemy
```

```python
sqlalchemy.__version__
```

```python
 from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///:memory:', echo=True)
```

```python
df.to_sql('sqladb', con=engine)
```

```python
engine.execute("SELECT age FROM sqladb").fetchall()
```

```python
engine.execute("UPDATE sqladb SET age=10 WHERE age>40")
engine.execute("SELECT age FROM sqladb").fetchall()
```

```python
engine.execute("UPDATE sqladb SET education=' NULL' WHERE sex=' Male'")
engine.execute("SELECT education,sex FROM sqladb").fetchall()
```

```python
engine.execute("DELETE FROM sqladb WHERE sex=' Male'")
engine.execute("SELECT sex FROM sqladb").fetchall()
```

```python
engine.execute("DELETE FROM sqladb WHERE age=40")
engine.execute("SELECT age FROM sqladb").fetchall()
```

```python
engine.execute("Select age From sqladb Where age>30").fetchall()
```

```python
engine.execute("Select age From sqladb Where sex=' Female'").fetchall()
```

```python
engine.execute("Select max(age) From sqladb").fetchall()
```

```python
engine.execute("Select avg(age) From sqladb").fetchall()
```
